### New Task: 20 Newsgroups + Vectorizer Comparison
Load a real text dataset and compare (in the shape) **Bag of Words (BOW)** with **Hashing Vectorizer**.

The **20 Newsgroups** dataset is a classic NLP benchmark made of posts from 20 discussion groups (topics).
In this task, we use only one topic (`sci.space`) so you can focus on how vectorization works.

```python
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups(subset='train', categories=['sci.space'])

texts = data.data[:100]  # Use first 100 documents
```

Meaning of the variables in this line:
- `data`: the dataset object returned by `fetch_20newsgroups(...)`. It contains fields such as `data` (documents), `target`, and `target_names`.
- `subset='train'`: load only the **training split** of the dataset.
- `categories=['sci.space']`: keep only documents from the **sci.space** newsgroup category.
- `texts`: a smaller list of documents selected from `data.data` (first 100 posts) to make experiments faster.

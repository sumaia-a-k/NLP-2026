# 1. Sentiment Analysis:
Sentiment analysis (or opinion mining) is the process of analyzing digital text to determine if the emotional tone of the message is positive, negative, or neutral. Some systems also detect more nuanced emotions like anger, joy, sadness, or frustration.

Today, companies have large volumes of text data like emails, customer support chat transcripts, social media comments, and reviews. Sentiment analysis tools can scan this text to automatically determine the author’s attitude towards a topic. Companies use the insights from sentiment analysis to improve customer service and increase brand reputation

The goal of sentiment analysis is to understand customer opinions, monitor brand reputation and improve business decisions based on real insights.

## Polarity and Subjectivity:

Polarity and subjectivity are two fundamental concepts in sentiment analysis, especially in text-based opinion mining. They help quantify and categorize the emotional content of a text.

- **Polarity**:It refers to the direction of the sentiment expressed in a text — essentially, whether the opinion is positive, negative, or neutral. It ranges between **[-1,1]**  where:

  
      - Positive (+1):	Favorable, happy, approving	, ex.:"I love this movie!"
      - Negative (-1):	Unfavorable, angry, critical, ex.:"This is the worst service ever."
      - Neutral (0):	Factual, no strong emotion,   ex.:"The store opens at 9 AM."


  

- **Subjectivity**: Measures how much the text expresses personal opinion, emotion, or judgment versus objective facts. Subjectivity ranges between **[0,1]** where:

      - High (close to 1):	Opinionated, emotional, personal, ex.: "This pizza tastes amazing!"

      - Low (close to 0):	Factual, neutral, objective, ex.:"The pizza contains cheese and tomato sauce."


  


### Relationship Between Polarity & Subjectivity:
- A highly subjective sentence usually has a clear polarity (positive or negative).

- A neutral polarity sentence can be either:

    - Objective (factual) → "The meeting is at 2 PM."

    - Subjective but balanced (mixed opinions) → "Some liked it, some hated it."
 
### Why Both Matter?
Polarity alone isn't enough — a factual news headline ("GDP fell by 2%") might have negative polarity but low subjectivity, so it shouldn't be treated as an angry opinion.

Subjectivity helps filter: For brand monitoring, you might only care about subjective sentences (actual opinions), ignoring neutral facts.

- ### Example:
  We're going to use `TextBlob` library, which uses a pre-trained sentiment prediction model.

```python
#import libraries
from textblob import TextBlob

# Initializing documents 
review = "I like this phone. screen quality and camera clarity is really good."
review2 = "This tv is not good. Bad quality, no clarity, worst experience"

#TextBlob has a pre trained sentiment prediction model
blob = TextBlob(review)
print("Sentiment for the first review: ",blob.sentiment,"\n")


#now lets look at the sentiment of review2
blob = TextBlob(review2)
print("Sentiment for the second review: ",blob.sentiment,"\n")
```

You can also get the get the values for polarity and subjectivity separately and in a more direct way by calling their attribute on the `blob` object:

```python
from textblob import TextBlob

text = "This movie is fantastic!"
blob = TextBlob(text)

print(blob.sentiment.polarity)    
print(blob.sentiment.subjectivity) 
```

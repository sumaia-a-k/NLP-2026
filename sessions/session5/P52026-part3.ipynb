# 1. Sentiment Analysis:
Sentiment analysis (or opinion mining) is the process of analyzing digital text to determine if the emotional tone of the message is positive, negative, or neutral. Some systems also detect more nuanced emotions like anger, joy, sadness, or frustration.

Today, companies have large volumes of text data like emails, customer support chat transcripts, social media comments, and reviews. Sentiment analysis tools can scan this text to automatically determine the author’s attitude towards a topic. Companies use the insights from sentiment analysis to improve customer service and increase brand reputation

The goal of sentiment analysis is to understand customer opinions, monitor brand reputation and improve business decisions based on real insights.

## Polarity and Subjectivity:

Polarity and subjectivity are two fundamental concepts in sentiment analysis, especially in text-based opinion mining. They help quantify and categorize the emotional content of a text.

- **Polarity**:It refers to the direction of the sentiment expressed in a text — essentially, whether the opinion is positive, negative, or neutral. It ranges between **[-1,1]**  where:

  
      - Positive (+1):	Favorable, happy, approving	, ex.:"I love this movie!"
      - Negative (-1):	Unfavorable, angry, critical, ex.:"This is the worst service ever."
      - Neutral (0):	Factual, no strong emotion,   ex.:"The store opens at 9 AM."


  

- **Subjectivity**: Measures how much the text expresses personal opinion, emotion, or judgment versus objective facts. Subjectivity ranges between **[0,1]** where:

      - High (close to 1):	Opinionated, emotional, personal, ex.: "This pizza tastes amazing!"

      - Low (close to 0):	Factual, neutral, objective, ex.:"The pizza contains cheese and tomato sauce."


  


### Relationship Between Polarity & Subjectivity:
- A highly subjective sentence usually has a clear polarity (positive or negative).

- A neutral polarity sentence can be either:

    - Objective (factual) → "The meeting is at 2 PM."

    - Subjective but balanced (mixed opinions) → "Some liked it, some hated it."
 
### Why Both Matter?
Polarity alone isn't enough — a factual news headline ("GDP fell by 2%") might have negative polarity but low subjectivity, so it shouldn't be treated as an angry opinion.

Subjectivity helps filter: For brand monitoring, you might only care about subjective sentences (actual opinions), ignoring neutral facts.

------------------------------------------

- ### Example:
  We're going to use `TextBlob` library, which uses a pre-trained sentiment prediction model.

```python
#import libraries
from textblob import TextBlob

# Initializing documents 
review = "I like this phone. screen quality and camera clarity is really good."
review2 = "This tv is not good. Bad quality, no clarity, worst experience"

#TextBlob has a pre trained sentiment prediction model
blob = TextBlob(review)
print("Sentiment for the first review: ",blob.sentiment,"\n")


#now lets look at the sentiment of review2
blob = TextBlob(review2)
print("Sentiment for the second review: ",blob.sentiment,"\n")
```

You can also get the values for polarity and subjectivity separately and in a more direct way by calling their attribute on your`blob` object:

```python
text = "This movie is fantastic!"
blob = TextBlob(text)

print("Polarity; ", blob.sentiment.polarity, "\n")    
print("Subjectivity :", blob.sentiment.subjectivity,"\n") 
```

# 2. Word Sense Disambiguation:

Word Sense Disambiguation (WSD) is a crucial task in NLP that aims to resolve this ambiguity by determining the correct sense of a word within a given context. It is the process of determining the intended meaning, or sense, of a word within a specific context. It involves mapping words to their appropriate sense in a given context to ensure accurate interpretation and understanding of text.

WSD is particularly important in applications such as machine translation, information retrieval, question answering systems, sentiment analysis, and natural language understanding.

## Lesk Algorithm:
Lesk Algorithm is a classical Word Sense Disambiguation algorithm introduced by Michael E. Lesk in 1986.

The Lesk algorithm is based on the idea that words in a given region of the text will have a similar meaning. In the Simplified Lesk Algorithm, the correct meaning of each word context is found by getting the sense which overlaps the most among the given context and its dictionary meaning.

 The Lesk algorithm is a dictionary-based approach that is considered seminal,it is earliest and still most commonly used methods for word-sense disambiguation

In the simplest terms, Lesk’s algorithm counts the number of overlaps between all dictionary definitions of a word of interest and all dictionary definitions of the words surrounding it, known as a “context window”. Then, it takes the definition corresponding to the word with the highest number of overlaps, without including stop words (words such as “the”, “a”, “and”), and infers it to be the word’s “sense”

Read More about the Lesk Algorithm [here](https://en.wikipedia.org/wiki/Lesk_algorithm).


### - Example:

We can use `NLTK` to implement Lesk in Python.

Let us start by downloading and importing the libraries:

```python
#Install pywsd
!pip install pywsd # pywsd library is a Python implementation of various Word Sense Disambiguation
```

```python
#Import functions
from nltk.corpus import wordnet as wn # WordNet = A lexical database (like a dictionary/thesaurus) that groups words into "senses" with definitions, synonyms, and relationships.
from nltk.stem import PorterStemmer
from itertools import chain
from pywsd.lesk import simple_lesk
```

```python
# Sentences
bank_sents = ['I went to the bank to deposit my money', 'The river bank was full of dead fishes']

# calling the lesk function and printing results for both the sentences
print ("Context-1:", bank_sents[0])
answer = simple_lesk(bank_sents[0],'bank')
print ("Sense:", answer)
print ("Definition : ", answer.definition())

print("\n", "**************", "\n")                     

print ("Context-2:", bank_sents[1])
answer = simple_lesk(bank_sents[1],'bank','n') # The 'n' parameter restricts search to noun senses only (bank can also be a verb: "to bank an airplane").
print ("Sense:", answer)
print ("Definition : ", answer.definition())
```

```python
a1= simple_lesk('This device is used to jam the signal','jam')
print(a1,a1.definition())

a2 = simple_lesk('I am stuck in a traffic jam','jam')
print(a2,a2.definition())

a3 = simple_lesk('I like strawberry jam','jam')
print(a3,a3.definition())
```

We notice that `simple_lesk` couldn't disambiguate the `jam` word in the 2nd and the 3rd sentences, So , on the following cell we're going to use the standard `lesk` to see the difference:

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

a1= lesk(word_tokenize('This device is used to jam the signal'),'jam')
print(a1,a1.definition())

a2 = lesk(word_tokenize('I am stuck in a traffic jam'),'jam')
print(a2,a2.definition())

a3 = lesk(word_tokenize('I like strawberry jam'),'jam')
print(a3,a3.definition())
```

- We see that `lesk` disambiguate the `jam` word proprely in the 3 sentences.
- So, the `NLTK` library can implement the Lesk method properly.

# 3. Speech Recognition and Speech to Text:

When people talk about converting audio into words, they often use speech recognition and speech-to-text interchangeably. While closely related, these two terms are not exactly the same — and understanding the difference can help you choose the right tool for your use case. This confusion is understandable because both technologies involve processing human speech. However, they serve different purposes and have distinct applications.


 ## Speech recognition:
is a broader technology that allows computers to identify and interpret human speech. It's an umbrella term that encompasses various applications where machines understand spoken language. The goal of speech recognition is **not only** to convert speech into text, but also to:
  
      - Understand commands — Process voice instructions and execute actions
      - Identify intent — Determine what the user wants to accomplish
      - Trigger actions — Perform tasks based on spoken input
      - Control systems — Interact with software, devices, or services
  
  Use Cases of Speech Recognition:
  
      - Voice assistants (Siri, Alexa, Google Assistant, Cortana)
      - Voice commands ("Turn on the lights", "Play music", "Set a timer")
      - Smart home devices (voice-controlled lights, thermostats, security systems)
      - Voice search (searching the web or apps using voice)
**Key point**: In many cases, speech recognition systems do not even display text to the user — the speech is simply analyzed and acted upon. The focus is on understanding intent and executing commands, not producing written transcripts.


## Speech-to-Text:
Speech-to-text (STT), also known as Automatic Speech Recognition (ASR) in transcription contexts, is a specific application of speech recognition focused on transcribing spoken language into written text. The primary goal of speech-to-text is:

    - Accuracy — Produce word-for-word accurate transcripts
    - Readability — Create clean, well-formatted text
    - Completeness — Capture everything that was said
    - Usability — Generate text that can be edited, searched, and shared
 Use Cases of Speech-to-Text:
 
     - Audio transcription — Convert recorded audio files to text
     - Subtitles and captions — Generate captions for videos and live streams
     - Content creation — Convert voice notes into written content
     - Podcast and interview transcripts — Create written records of conversations
     - Accessibility — Provide text alternatives for audio content

**Key point**: If your main need is to turn audio or video files into text, then speech-to-text is exactly what you're looking for. The output is always text that you can read, edit, and use in other applications.

  

- ### Example:

```python
#Install and import necessary libraries
!pip install SpeechRecognition
!pip install PyAudio
import speech_recognition as sr
```

```python
r=sr.Recognizer()

# sr.Microphone accesses your computer's microphone as an audio source
#  Since we used the 'with' block, it opens the microphone and automatically closes it afterward
with sr.Microphone() as source: 
    print("Please say something")
    audio = r.listen(source, phrase_time_limit=5)  # Records audio until silence is detected (or 1-2 seconds of no speech)
    print("Time over, thanks")
try:
    print("I think you said: "+r.recognize_google(audio)); # Sends the recorded audio to Google Web Speech API (free, requires internet)
except:                                                    # Returns the transcribed text
    pass;
```

### Using another language (German):

```python
r=sr.Recognizer()
with sr.Microphone() as source:
    print("Please say something")
    audio = r.listen(source, phrase_time_limit=5)
    print("Time over, thanks")
try:
    print("I think you said: "+r.recognize_google(audio, language ='de-DE'));
except sr.UnknownValueError:
    print("Google Speech Recognition could not understand audio")
except sr.RequestError as e:
    print("Could not request results from Google Speech Recognition service; {0}".format(e))
except:
    pass;
```

# BM25: A Retrieval-Oriented Ranking Algorithm

BM25 (Best Matching 25) is one of the most widely used ranking algorithms in Information Retrieval systems such as search engines.

It was designed to improve document ranking quality compared with simpler term-weighting approaches by handling term frequency and document length in a more retrieval-focused way.

## Why Do We Need BM25?

BM25 builds on the traditional TF-IDF approach but introduces two critical improvements:

- Term Frequency (TF) Saturation: In older models, a word appearing 100 times might score 100 times higher than a word appearing once. BM25 uses a "saturation" curve where the benefit of seeing a word again diminishes after a few occurrences, preventing "keyword stuffing" from unfairly boosting scores.

- Document Length Normalization: It recognizes that longer documents naturally contain more words. It penalizes long, rambling documents and rewards shorter, more concise ones that contain the same query terms, assuming they are more likely to be specifically about that topic. 


## BM25 Scoring Intuition

For a query term $t$ and document $d$, BM25 uses an IDF-like weight plus a saturated TF component. A common form is:

$$ BM25(t, d) = IDF(t) \cdot \frac{f(t,d) \cdot (k_1 + 1)}{f(t,d) + k_1 \cdot (1 - b + b \cdot \frac{|d|}{avgdl})} $$

Where:

- $f(t,d)$: term frequency of $t$ in $d$
- $|d|$: document length
- $avgdl$: average document length in corpus
- $k_1$: controls TF saturation (typical range about 1.2 to 2.0)
- $b$: controls length normalization strength (often around 0.75)

understanding more the core parameters:


1- k1(Term Frequency Saturation)
This parameter controls how much "credit" a document gets for repeating a keyword.

The Problem: In basic systems, if "pizza" appears 20 times, the score is 20x higher than if it appears once. This is bad because a document that mentions "pizza" 20 times isn't necessarily 20 times more relevant than one that mentions it 3 or 4 times.

The

Fix: It creates a "plateau." As the word appears more often, the score increases quickly at first, then slows down.
- Low
    (e.g., 1.2): The score levels off very quickly. Whether a term appears 3 times or 30 times, the score will be nearly the same. This is good if you want to prevent "keyword stuffing."
- High
    (e.g., 2.0): The score keeps climbing for longer. Use this if your documents are very long or technical and you expect relevant terms to appear frequently.

2- b(Length Normalization)
This parameter accounts for the fact that a long book is naturally more likely to contain a word than a short tweet, simply because it has more words.

The Problem: Without normalization, long documents almost always rank higher than short ones because they have more opportunities to match keywords.
The Fix: It adjusts the score based on how the document's length compares to the average length of all documents in your collection.

- b=0 (No normalization): Length is ignored. A 500-page book and a 1-sentence headline are treated equally. This is rarely ideal.
        
- b=1 (Full normalization): The score is heavily penalized if the document is longer than average. This assumes that if a long document and a short document both mention "pizza" twice, the short one is more "about" pizza.


## Main Differences: BM25 vs TF-IDF

| Aspect | TF-IDF | BM25 |
|---|---|---|
| Goal | General term weighting / vector features | Retrieval ranking optimized for search |
| TF behavior | Usually linear or normalized by document length | Saturates with diminishing returns |
| Typical use | Feature extraction, classification, clustering | Query-document ranking in search engines |
| Practical ranking quality | Good baseline | Often stronger for retrieval tasks |

## When To Use Which

- Use TF-IDF when you need simple, interpretable text features for downstream ML tasks.
- Use BM25 when your main objective is ranking documents for search relevance.

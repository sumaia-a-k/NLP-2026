# TF-IDF Feature Extraction



In the realm of Natural Language Processing (NLP), **TF-IDF** (Term Frequency-Inverse Document Frequency) is a powerful technique used to analyze and understand the importance of words in a document corpus. TF-IDF plays a crucial role in tasks such as text mining, information retrieval, document ranking, keyword extraction, and document classification.



The main idea behind TF-IDF is simple: a word should be considered important if it appears frequently in a specific document, but not too frequently across all documents. This balance helps us identify terms that are more meaningful for describing the content of a document.


## What Is TF-IDF?



TF-IDF (Term Frequency-Inverse Document Frequency) is a statistical measure used in natural language processing and information retrieval to evaluate the importance of a word in a document relative to a collection of documents called a **corpus**.



TF-IDF balances common and rare words to highlight the most meaningful terms. A term that appears many times in one document may be important for that document, but if the same term appears in nearly every document, then it is less useful for distinguishing one document from another.



In practice, TF-IDF is often used to:



- represent text numerically for machine learning models

- identify keywords in a document

- compare documents based on important terms

- support search engines and information retrieval systems



## Term Frequency: TF(t, d)



Term Frequency (TF) measures how often a term $t$ appears in a document $d$. The intuition is that a term appearing frequently in a document is likely to be relevant to that document's content.



The standard formula is:



$$

TF(t, d) = \frac{\text{number of times term } t \text{ appears in document } d}{\text{total number of terms in document } d}

$$



A higher TF value means the term is more prominent within that specific document.



### Limitation of TF Alone



TF by itself does not consider how common the term is across the entire corpus. For example, words such as "the", "is", or "and" may appear frequently in many documents. These words can have high TF values, but they usually do not help distinguish one document from another.



## Inverse Document Frequency: IDF(t)



Inverse Document Frequency (IDF) measures how unique or rare a term is across the corpus. If a term appears in many documents, its IDF should be low. If it appears in only a few documents, its IDF should be high.



A common form of the IDF formula is:



$$

IDF(t) = \log\left(\frac{N}{df(t)}\right)

$$



Where:



- $N$ is the total number of documents in the corpus

- $df(t)$ is the number of documents containing the term $t$



### Why Do We Use the Log in IDF?



The logarithm is used to **compress the scale** of the IDF values. Without the log, very rare words could receive extremely large weights, which would make the feature values less stable and less balanced.



We usually want IDF to increase when a word becomes rarer, but we do not want that increase to be too aggressive. Using the logarithm makes IDF grow **sublinearly** with rarity.



For example, suppose we have a corpus of $1000$ documents:



- if a term appears in $100$ documents, then $\frac{N}{df(t)} = \frac{1000}{100} = 10$

- if a term appears in $10$ documents, then $\frac{N}{df(t)} = \frac{1000}{10} = 100$

- if a term appears in $1$ document, then $\frac{N}{df(t)} = \frac{1000}{1} = 1000$



Without the logarithm, the IDF-related values would be $10$, $100$, and $1000$, so the scale grows very quickly.



If we use the **natural logarithm** $\ln$, then:



$$

\ln(10) \approx 2.30, \qquad \ln(100) \approx 4.61, \qquad \ln(1000) \approx 6.91

$$



Now the values are still increasing as the term becomes rarer, but they increase much more gradually than $10$, $100$, and $1000$.



This gives us two important benefits:



- rare words still receive higher importance

- extremely rare words do not dominate the representation too much



So the log helps keep TF-IDF more balanced and more useful as a numerical feature representation.



In practice, different libraries may use slightly different IDF variants, and the log base can vary, but the main idea remains the same: the logarithm keeps the growth controlled.



This means rare terms receive higher weights, while very common terms receive lower weights.



## Combining TF and IDF



TF-IDF combines these two ideas into a single score:



$$

TF\text{-}IDF(t, d) = TF(t, d) \times IDF(t)

$$



This product gives a high score to words that:



- appear frequently in one document

- appear in relatively few documents across the corpus



As a result, TF-IDF highlights words that are both **locally important** and **globally informative**.



## Why TF-IDF Is Useful for Feature Extraction



When building NLP systems, raw word counts are often not enough. TF-IDF transforms text into weighted numerical features that better capture the significance of each word. These features can then be used in tasks such as:



- document classification

- clustering

- similarity measurement

- search and ranking




## When TF-IDF Is Still Useful



Even though modern NLP systems often use embeddings and large language models, TF-IDF still has important practical advantages in many real-world tasks.



- **Exact keyword matching:** TF-IDF works very well when the exact word or token matters. For example, if we search for a specific error code such as `0x8004210B`, TF-IDF can rank documents containing that exact token very effectively, while semantic models may focus more on the overall meaning of the text.

- **Speed and cost:** TF-IDF is computationally lightweight. It can be calculated quickly on a CPU and is often much cheaper than neural retrieval systems when working with very large document collections.

- **No pretraining required:** TF-IDF is a statistical method, so it can be applied immediately to a new corpus without pretraining or domain adaptation.

- **Strong baseline:** In many information retrieval and text classification tasks, TF-IDF remains a strong baseline because it is simple, interpretable, and often surprisingly effective.



This does not mean TF-IDF is always better than modern neural methods. Instead, it means TF-IDF is still a very useful tool, especially when we need simplicity, efficiency, and exact term matching.

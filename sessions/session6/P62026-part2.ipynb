```python
def read_file(filePath):
    with open(filePath) as f:
        str_text = f.read()

    return str_text
```

```python
read_file('moby_dick_four_chapters.txt')
```

```python
# tokenize and clean the text 
import spacy
```

```python
nlp = spacy.load("en_core_web_sm" , disable = ['parser', 'tagger' , 'net'])
```

```python
nlp.max_length = 1198623 # sometimes we need to increase it just in case
```

```python
def separate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
d = read_file('moby_dick_four_chapters.txt')
tokens = separate_punc(d)
```

```python
len(tokens)
```

```python
# we need to create sequence of tokens 
# we are going to pass first 49 words of the sentence and let our network predict the next word 
# 49 word ==> network predict #50
train_len = 49+1 
text_sequences = []

for i in range(train_len, len(tokens)):
    
    # Grab train_len# amount of characters
    seq = tokens[i-train_len:i]
    
    # Add to list of sequences
    text_sequences.append(seq)
```

```python
# keras tokenization to format this in numeric system that keras understands
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_sequences)
sequences = tokenizer.texts_to_sequences(text_sequences)
```

```python
X = sequences[:,:-1]
y = sequences[:,-1]
```

```python
seq_len = X.shape[1]
print(seq_len)
```

```python
# change Y to catgegory 
y = to_categorical(y, num_classes=vocabulary_size+1) # +1 for the padding zeros
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding
```

```python
def create_model(vocabulary_size, sl):
    model = Sequential()
    # input dim = vocabulary_size , output_dim = sequnce length (sl)
    model.add(Embedding(vocabulary_size, sl)) # has to be the first layer in the model
    model.add(LSTM(150, return_sequences=True)) 
    model.add(LSTM(150))
    model.add(Dense(150, activation='relu'))

    model.add(Dense(vocabulary_size, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    return model
```

```python
model = create_model(vocabulary_size+1, seq_len)
```

```python
from pickle import dump,load 
model.fit(X, y, batch_size=128, epochs=10,verbose=1)
```

```python
model.summary()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    '''
    INPUTS:
    model : model that was trained on text data
    tokenizer : tokenizer that was fit on text data
    seq_len : length of training sequence
    seed_text : raw string text to serve as the seed
    num_gen_words : number of words to be generated by model
    '''
    
    # Final Output
    output_text = []
    
    # Intial Seed Sequence
    input_text = seed_text
    
    # Create num_gen_words
    for i in range(num_gen_words):
        
        # Take the input text string and encode it to a sequence
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        
        # Pad sequences to our trained rate 49
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        
        # Predict raw probabilities for all words in vocabulary
        raw_predictions = model.predict(pad_encoded, verbose=0)[0]

        # Extract index of the highest probability word using argmax
        pred_word_ind = np.argmax(raw_predictions)

        # Grab word
        pred_word = tokenizer.index_word[pred_word_ind] 
        
        # Update the sequence of input text (shifting one over with the new word)
        input_text += ' ' + pred_word
        
        output_text.append(pred_word)
        
    # Make it look like a sentence.
    return ' '.join(output_text)
```

```python
seed_text = "Call me drawn toward the fathomless ocean whenever the damp November of my soul grows cold. There sails the grand phantom of the whale through stormy waters, a majestic creature pursued by a mad captain across the rolling tides of the vast Pacific, seeking vengeance upon the great white monster" 
```

```python
generate_text(model,tokenizer,seq_len,seed_text=seed_text,num_gen_words=10)
```

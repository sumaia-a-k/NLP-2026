## First Exercise:
you have the following text: `Text = "NLP deals with several methods and techniques to preprocess the text data along with exploratory data analysis. NLP -> natural language processing"`.

### Task:
- **Count the number of word which have the base "Process".**

## Solution:
We're going to use Regex (Regular Expressions) to achieve the task:

```python
import re # Importin Regex library

Text = """    NLP deals with several methods and techniques to preprocess the text data along with exploratory data 
    analysis. NLP -> natural language processing
    
    """


# Cleaning the text and splitting it into words
# \w = any “word character” (letters, digits, underscore), roughly [A-Za-z0-9_] in basic cases.
words = re.findall(r'\w+', Text.lower()) # We're using lowercase to ensure that we don't miss "Process" vs "process"

# Counting the words that contain the substring "process"
base = "process"
matching_words = [word for word in words if base in word]
count = len(matching_words)

# Printing the results
print(f"Occurrences found: {count}")
print(f"Words matched: {matching_words}")
```

```python
# another way 
import re
from nltk.stem import PorterStemmer

Text = """NLP deals with several methods and techniques to preprocess the text data along with exploratory data analysis. NLP -> natural language processing"""

# 1) tokenize
words = re.findall(r'\w+', Text.lower())

# 2) stem
stemmer = PorterStemmer()
stems = [stemmer.stem(w) for w in words]
# print stems to see how they look like
print(stems)

# Target base
target_stem = stemmer.stem("process")  # "process"

# Strict match: only words whose stem is exactly "process"
strict_matches = [w for w in words if stemmer.stem(w) == target_stem]
print("Strict count:", len(strict_matches))
print("Strict matches:", strict_matches)

# Optional relaxed match: include prefixed forms like "preprocess"
relaxed_matches = [w for w in words if target_stem in stemmer.stem(w)]
print("Relaxed count:", len(relaxed_matches))
print("Relaxed matches:", relaxed_matches)
```

## Second Exercise:
you have the following text: `Text = "NLP deals with several methods and techniques to preprocess the text data along with exploratory data analysis."`.

### Task:
- **Extract the most important words in order to use in a search task.**

## Solution:
We're going to use:
- `stopwords`from `nltk`: for removing stop words.
- `word_tokenize`: for tokenizing into words.
  
in order to achieve the task:

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

Text = "NLP deals with several methods and techniques to preprocess the text data along with exploratory data analysis."

# Tokenize and Lowercase
words = word_tokenize(Text.lower())

# Filter out stop words and punctuation
stop_words = set(stopwords.words('english'))
important_words = [w for w in words if w not in stop_words]

print(important_words)
```

## Third Exercise:
You have the following text: `Text = ['This is introduction to NLP','It is likely to be useful, to people ','Machine learning is the new electrcity', 'There would be less hype around AI and more action going forward','python is the besttool!','R is good langauage', 'I like this book','I want more books like this']`.

### Task:
- **Determine the number of words in this sentence .**
- **Print the most repeated word while considering the words roots (bases).**

## Solution:
We're going to use:
- `WordNetLemmatizer`from `nltk.stem` : to determine the number of words and find the most repeated root.
- `Counter` class from the `collections` module : Only for the sake of not needing to write a very complex `for` loop. *just for simplicity.*
  
in order to achieve the task:

```python
from nltk.stem import WordNetLemmatizer
from collections import Counter
# Counter is a class in Python standard library that counts how many times each item appears in an iterable.
from nltk.corpus import stopwords
# Initialize a lemmatizer object
lemmatizer = WordNetLemmatizer()


Text = [
    'This is introduction to NLP',
    'It is likely to be useful, to people ',
    'Machine learning is the new electrcity', 
    'There would be less hype around AI and more action going forward',
    'python is the besttool!',
    'R is good langauage', 
    'I like this book',
    'I want more books like this'
]

# Filter out stop words and punctuation
stop_words = set(stopwords.words('english'))

# Preprocessing the text:removing punctution, removing stopwords and start Lemmatizing
all_roots = []

for sentence in Text:
    # Removing punctuation and starting splitting
    words = re.findall(r'\w+', sentence.lower())
    for word in words:
        if word not in stop_words:  # Removing stop words
            # Finding the root 
            root = lemmatizer.lemmatize(word)
            all_roots.append(root)
```

```python
# starting to analyse
total_word_count = len(all_roots) # Counting the number of the roots that we have

root_counts = Counter(all_roots) # Defining a counter object.

# returns a list containing the top 1 most frequent item as a tuple: [(word, count)]
most_common_root, frequency = root_counts.most_common(1)[0] # Finding the most common word

print(f"Total processed words (excluding stop words): {total_word_count}")
print(f"Most repeated root: '{most_common_root}' (appears {frequency} times)")
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, HashingVectorizer
from nltk import bigrams
import nltk

feedback = [
    "The food was great and the service was fast.",
    "The service was slow but the food was delicious.",
    "Great food and amazing service."
]
```

### Using BiGram:

```python
cv = CountVectorizer(ngram_range=(2, 2), stop_words='english')
bigram_counts = cv.fit_transform(feedback)

print(cv.vocabulary_,"\n")
print(bigram_counts.toarray())
```

### Using Co-Occurrence Matrix:

```python
tokens = " ".join(feedback).lower().split()
vocab = sorted(list(set(tokens)))
word_to_id = {word: i for i, word in enumerate(vocab)}

# Building the matrix
co_matrix = np.zeros((len(vocab), len(vocab)))
for w1, w2 in bigrams(tokens):
    co_matrix[word_to_id[w2]][word_to_id[w1]] += 1

df_co = pd.DataFrame(co_matrix, index=vocab, columns=vocab)

print(df_co,"\n")
print (df_co.shape)
```

### Using Hash-Vectorizer

```python
hv = HashingVectorizer(n_features=5)
hashed_output = hv.transform(feedback)

print(hashed_output.toarray(),"\n")
print (hashed_output.shape)
```

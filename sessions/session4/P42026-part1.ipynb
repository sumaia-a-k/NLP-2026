# Introduction to Text Similarity

Text similarity is the process of comparing two pieces of text and measuring how close they are. In simple terms, it tells us how much two words, sentences, or documents resemble each other. Because this task is centered on language, it is a core topic in Natural Language Processing (NLP).

In many real-world applications, we need to compare texts based on meaning, not only exact word overlap. For example:

- Search engines estimate how relevant a document is to a query, even when the same words are not used.
- Question-and-answer platforms such as Quora or Stack Overflow detect whether a similar question has already been asked.
- E-commerce systems recommend the closest available product when the exact item is out of stock.
- Organizations compare documents, reports, or letters to identify duplicates or near-duplicates.
- Recruitment systems match resumes to the most suitable job role or profile.

In this session, we will study common ways to measure similarity, including lexical methods (word overlap) and semantic methods (meaning overlap).

1- Cosine Similarity 

Cosine similarity measures the similarity between two vectors of an inner product space in general. It measures the cosine of the angle between two embeddings and determines whether they are pointing in roughly the same direction or not. When the embeddings are pointing in the same direction the angle between them is zero so their cosine similarity is 1. When the embeddings are perpendicular to each other the angle between them is 90 degrees and the cosine similarity is 0 finally when the angle between them is 180 degrees the cosine similarity is -1.


similarity(A, B) =  
$$\frac{A \cdot B}{\|A\| \times \|B\|} = \frac{\sum_{i=1}^n A_i \times B_i}{\sqrt{\sum_{i=1}^n A_i^2} \times \sqrt{\sum_{i=1}^n B_i^2}}$$


Let us say v1 and v2 are vectors.

v1 = [ 1 3 2 ]

v2 = [ 5 0 -3]

In this case, the dot product of v1 and v2 will be:

v1 . v2 = 1 * 5 + 3 * 0 + 2 * -3 = 5 + 0 + -6 = -1

Magnitude will be:

|v1| = 1² + 3² + 2² = 1 + 9 + 4 = 14

|v2| = 5² + 0² + (-3)² = 25 + 0 + 9 = 34

The cosine similarity is: $\frac{-1}{\sqrt{14} \times \sqrt{34}} \approx -0.0458$


2- Jaccard Similarity

Jaccard Similarity (Intersection over Union) is defined as:

J(A, B) = |A ∩ B| / |A ∪ B|

This standard form is for sets (for example, sets of unique words).
For weighted/count vectors, use Weighted Jaccard instead.

Example using sets of words (not vectors):

A = {cat, mat, sat, the, on}
B = {dog, mat, sat, the, on}

A ∩ B = {the, sat, on, mat}

Size of intersection = 4
Size of union = 6

$$
\text{Jaccard}(A,B) = \frac{|A \cap B|}{|A \cup B|} = \frac{4}{6} \approx 0.667
$$

3- Weighted Jaccard Similarity

In standard Jaccard, two documents are compared as simple sets of unique words, so word frequency does not affect the score. Weighted Jaccard extends this idea by comparing nonnegative feature vectors (for example TF or TF-IDF) index by index.

$$J_w(\mathbf{x}, \mathbf{y}) = \frac{\sum_{i} \min(x_i, y_i)}{\sum_{i} \max(x_i, y_i)}$$

Let:
d1 = [1, 3, 2]
d2 = [5, 0, 3]

Compute by index:
- min vector = [min(1,5), min(3,0), min(2,3)] = [1, 0, 2], so numerator = 3
- max vector = [max(1,5), max(3,0), max(2,3)] = [5, 3, 3], so denominator = 11

So Weighted Jaccard similarity = 3/11 ≈ 0.273


3- Euclidean Distance

The Euclidean Distance formula is the most common formula to calculate distance between two points/coordinates. To get it we just need to subtract the points from the vectors, raise them to squares, add them up and take the square root of them. Similarly, it’s used to calculate the distance between the embedding vectors of the words, supposing each vector represents a set of coordinates.

$$
d(p, q) = d(q, p) = \sqrt{(q_1 - p_1)^2 + (q_2 - p_2)^2 + \cdots + (q_n - p_n)^2}
$$
$$
= \sqrt{\sum_{i=1}^n (q_i - p_i)^2}
$$

v1 = [ 1 3 2 ]

v2 = [ 5 0 -3]

The Euclidean distance will be:

Euclidean distance = sqrt ( (1–5)² + (3–0)² + (2-(-3))² )

= sqrt ( 16 + 9 + 25 )

= sqrt (50)

= 5 * sqrt(2)

Understanding Lexical Similarity and Semantic Similarity

When we are working on similarity projects, we mainly need to compute how ‘close’ two pieces of text are, either in meaning(semantic approach) or in surface closeness(Lexical approach).

Lexical Similarity:

For example if we compare the sentence: The cat ate the mouse.; with the sentence: The mouse ate the cat.; the similarity score will be very high as the words in both the sentences are almost same whatever meaning it makes.

example:
Keywords matching — Selecting texts based on given keywords like finding resumes with similar skills set keywords or job role keywords.


Semantic Similarity:

Semantic similarity refers to the similarity of two pieces of text when their contextual meaning is considered. It judges the order of occurrences of the words in the text.

For example, if there are two sentences: The horse is carrying the load; and the load is carrying the horse, where we try to find the semantic similarity score between them then the score will be very less

-Lexical asks “Do the words look alike?”

-Semantic asks “Do the meanings match?”

2- Compute IDF: 

IDF(affection) = $\ln\left(\frac{4}{1+3}\right) + 1 = \ln(1) + 1 = 1$

IDF(Jealous)   = $\ln\left(\frac{4}{1+3}\right) + 1 = \ln(1) + 1 = 1$ 

IDF(Gossip)    = $\ln\left(\frac{4}{1+2}\right) + 1 = \ln\left(\frac{4}{3}\right) + 1 \approx 0.2877 + 1 = 1.2877$

IDF(weathering)= $\ln\left(\frac{4}{1+1}\right) + 1 = \ln\left(\frac{4}{2}\right) + 1 = \ln(2) + 1 \approx 0.6931 + 1 = 1.6931$


$$
115^2 + 10^2 + 2.575^2 + 0^2 = 13225 + 100 + 6.631 = 13331.631

\sqrt{13331.631} \approx 115.46
$$

$$
\left[ \frac{115}{115.46}, \frac{10}{115.46}, \frac{2.575}{115.46}, \frac{0}{115.46} \right]
\approx [0.996, 0.0866, 0.0223, 0]
$$

**2. Novel2:**

$$
58^2 + 7^2 + 0^2 + 0^2 = 3364 + 49 = 3413

$$
\sqrt{3413} \approx 58.42

$$
\left[ \frac{58}{58.42}, \frac{7}{58.42}, \frac{0}{58.42}, \frac{0}{58.42} \right]
\approx [0.993, 0.1198, 0, 0]
$$

**3. Novel3:**

Sum of squares:

### Cosine Similarity Between Novels

Since each row vector is L2‑normalized, the cosine similarity is exactly the dot product.

**(1) Novel1 vs. Novel2**
$$
\begin{aligned}
\text{sim} &= (0.996 \times 0.993) + (0.0866 \times 0.1198) + (0.0223 \times 0.0) + (0.0 \times 0.0) \\
&= 0.989028 + 0.01037468 + 0 + 0 \\
&= 0.99940268 \\
&\approx 0.9994
\end{aligned}
$$

**(2) Novel1 vs. Novel3**
$$
\begin{aligned}
\text{sim} &= (0.996 \times 0.812) + (0.0866 \times 0.4466) + (0.0223 \times 0.3137) + (0.0 \times 0.2062) \\
&= 0.808752 + 0.03867556 + 0.00699551 + 0 \\
&= 0.85442307 \\
&\approx 0.8544
\end{aligned}
$$

**(3) Novel2 vs. Novel3**
$$
\begin{aligned}
\text{sim} &= (0.993 \times 0.812) + (0.1198 \times 0.4466) + (0.0 \times 0.3137) + (0.0 \times 0.2062) \\
&= 0.806316 + 0.05350268 + 0 + 0 \\
&= 0.85981868 \\
&\approx 0.8598
\end{aligned}


```
**Summary table:**

| Pair             | Cosine Similarity |
|------------------|-------------------|
| Novel1 – Novel2  | 0.9994            |
| Novel1 – Novel3  | 0.8544            |
| Novel2 – Novel3  | 0.8598            |
```

````python
documents = ( "I like NLP", "I am exploring NLP", "I am a beginner in NLP", "I want to learn NLP", "I like advanced NLP")
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

#compute similarity for first sentence with rest of the sentences
cosine_similarity(tfidf_matrix[0:1],tfidf_matrix)
````

````python
from sklearn.metrics.pairwise import pairwise_distances

# First binarize the TF-IDF matrix (word presence/absence)
# The tfidf_matrix contains continuous values (weights). Standard Jaccard is a set-based metric that only cares if a word exists or not.
# (tfidf_matrix > 0) converts every TF-IDF weight into True (word exists) or False (word doesn't exist).
binary_matrix = (tfidf_matrix > 0) # this the standard jaccard since we are converting it to binary 
#Converts the SciPy sparse matrix into a dense NumPy array.
binary_matrix = binary_matrix.toarray()

# Compute Jaccard *distance* between first doc and all others
jaccard_dists = pairwise_distances(binary_matrix[0:1], binary_matrix, metric='jaccard')
# Convert to similarity
# subtracting the result from 1 flips the scale - since it is distance
jaccard_binary_sim = 1 - jaccard_dists

print(jaccard_binary_sim)
````

### Numeric example: cosine similarity using word embeddings

We will first compute the sentence embedding for each sentence by averaging the word vectors, then compute cosine similarity between the sentence vectors.

Sentences:
- S1 = I love reading books
- S2 = I like writing novels
- S3 = I like playing with cats

Word embeddings:

#### Step 1: sentence embeddings

For S1 = I love reading books

$$
S_1 = \frac{1}{4} \Big([0.02,0.01,0.00,0.95] + [0.87,0.42,0.13,0.05] + [0.11,0.88,0.22,0.02] + [0.06,0.10,0.92,0.03]\Big)

$$
S_1 = \frac{1}{4}[1.06,1.41,1.27,1.05] = [0.265, 0.3525, 0.3175, 0.2625]
$$

For S2 = I like writing novels

$$
S_2 = \frac{1}{4} \Big([0.02,0.01,0.00,0.95] + [0.65,0.38,0.10,0.04] + [0.09,0.85,0.19,0.02] + [0.07,0.12,0.89,0.03]\Big)
$$

$$
S_2 = \frac{1}{4}[0.83,1.36,1.18,1.04] = [0.2075, 0.34, 0.295, 0.26]
$$

For S3 = I like playing with cats

$$
S_3 = \frac{1}{5} \Big([0.02,0.01,0.00,0.95] + [0.65,0.38,0.10,0.04] + [0.14,0.91,0.25,0.01] + [0.00,0.01,0.03,0.78] + [0.18,0.08,0.85,0.02]\Big)
$$

$$
S_3 = \frac{1}{5}[0.99,1.39,1.23,1.80] = [0.198, 0.278, 0.246, 0.36]
$$

#### Step 2: cosine similarity formula

$$
\cos(\theta) = \frac{A \cdot B}{\|A\|\|B\|}
$$

#### Step 3: similarity between S1 and S2

Dot product:

$$
S_1 \cdot S_2 = (0.265)(0.2075) + (0.3525)(0.34) + (0.3175)(0.295) + (0.2625)(0.26)
$$

$$
S_1 \cdot S_2 = 0.0550 + 0.1199 + 0.0937 + 0.0683 = 0.3369
$$

Norms:

$$
\|S_1\| = \sqrt{0.265^2 + 0.3525^2 + 0.3175^2 + 0.2625^2} \approx 0.6026
$$

$$
\|S_2\| = \sqrt{0.2075^2 + 0.34^2 + 0.295^2 + 0.26^2} \approx 0.5605
$$

Cosine similarity:

$$
\cos(S_1,S_2) = \frac{0.3369}{(0.6026)(0.5605)} \approx 0.9970
$$

#### Step 4: remaining similarities

$$
\cos(S_1,S_3) \approx 0.9668
$$

$$
\cos(S_2,S_3) \approx 0.9737
$$

| Pair | Cosine similarity |
|---|---:|
| S1 and S2 | 0.9970 |
| S1 and S3 | 0.9668 |
| S2 and S3 | 0.9737 |

### Practical code example using `KeyedVectors`

This version uses `gensim.models.KeyedVectors` to store the word embeddings, creates sentence embeddings with mean pooling, and then computes cosine similarity between two sentences.

````python
import numpy as np
from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity

# 1. Store the given word embeddings in a dictionary
data = {
    "i": np.array([0.02, 0.01, 0.00, 0.95]),
    "love": np.array([0.87, 0.42, 0.13, 0.05]),
    "like": np.array([0.65, 0.38, 0.10, 0.04]),
    "reading": np.array([0.11, 0.88, 0.22, 0.02]),
    "writing": np.array([0.09, 0.85, 0.19, 0.02]),
    "playing": np.array([0.14, 0.91, 0.25, 0.01]),
    "books": np.array([0.06, 0.10, 0.92, 0.03]),
    "novels": np.array([0.07, 0.12, 0.89, 0.03]),
    "cats": np.array([0.18, 0.08, 0.85, 0.02]),
    "with": np.array([0.00, 0.01, 0.03, 0.78]),
}

# 2. Convert the dictionary to a KeyedVectors object
kv = KeyedVectors(vector_size=4) # matches the embedding dimension.
kv.add_vectors(list(data.keys()), list(data.values())) # bulk adds words and their vectors.

# 3. Create a sentence embedding by averaging all word vectors in the sentence
def get_sentence_embedding(sentence, model):
    """Return the mean-pooled embedding for a sentence."""
    words = sentence.lower().split() # Lowercases the sentence to match dictionary keys (all lowercase).
    vectors = [model[word] for word in words if word in model] # Retrieves vectors only for words present in the model (ignores unknown words).
    if not vectors:
        raise ValueError("No words from the sentence were found in the embedding model.")
    return np.mean(vectors, axis=0)

# 4. Generate sentence vectors
s1_vec = get_sentence_embedding("I love reading books", kv)
s2_vec = get_sentence_embedding("I love playing with cats", kv)

# 5. Display the vectors and compute cosine similarity
print(f"Sentence 1 Vector: {np.round(s1_vec, 4)}")
print(f"Sentence 2 Vector: {np.round(s2_vec, 4)}")

# reshape(1, -1) converts it into a 2D array with 1 row and as many columns as needed
# s1_vec.shape   # (4,)
# s1_vec.reshape(1, -1).shape   # (1, 4)
# cosine_similarity from sklearn.metrics.pairwise expects 2D arrays – specifically, each input should be of shape (n_samples, n_features).
score = cosine_similarity(s1_vec.reshape(1, -1), s2_vec.reshape(1, -1))[0][0]
print(f"Cosine Similarity Score: {score:.4f}")
````

# Noun-Phrase-Extraction:

Noun phrase (NP) extraction (also called NP-chunking) is an NLP technique that identifies and extracts nouns and their modifiers (adjectives, determiners) from text to capture key information - who, what, and which specific things are being discussed -  often using libraries like `SpaCy`, `NLTK`, or `TextBlob`. 

- It transforms raw text into structured data by locating chunks such as: "the startup" a known/specific startup vs "a strat up" any start up
- Often focusing on adjective-noun or noun-noun patterns. 
- It is a fundamental task in Natural Language Processing (NLP) used for summarization, information retrieval, Question answering and content understanding. 

Noun phrases are POS patterns that include one or multiple nouns.

Common Patterns of NP: Phrases are identified by specific part-of-speech (POS) sequences, such as:

- Adjective + Noun: "Beautiful text"
- Noun + Noun: "Machine Learning"
- Determiner + Noun: "The cat"
- Determiner + Adjective+  Noun: " The lazy dog".

- ### Using `NLTK`:
NLTK requires a manual approach. WE must first tag the words with Part-of-Speech (POS) tags and then define a Regular Expression grammar to "chunk" them together.


```python
import nltk
from nltk import pos_tag, word_tokenize, RegexpParser

text = "The quick brown fox jumps over the lazy dog near the old bridge."

# 1. Tokenize and POS-tag each token
tokens = word_tokenize(text)
tags = pos_tag(tokens)
print("Tokens:", tokens)
print("POS tags:", tags)

# 2. Define NP grammar: optional determiner + zero/more adjectives + singular noun
# <Chunk_Label>: { <POS_Pattern> }
# NP: → Tells the parser: "When you find a match, wrap those tokens in a tree node labeled NP."
grammar = "NP: {<DT>?<JJ>*<NN>}"
print("\nGrammar rule:", grammar)

# 3. Build the chunk parser and parse the tagged sentence
parser = RegexpParser(grammar)
tree = parser.parse(tags)

# What does parse() do?
# parse(tags) applies the grammar to POS-tagged tokens and returns an nltk.Tree object.
# This tree contains normal tagged tokens and grouped NP chunks as Tree nodes.
print("\nType returned by parse():", type(tree))
print("Full parse tree:")
print(tree)

# What is tree.subtrees()?
# tree.subtrees() returns a generator over EVERY subtree in the tree (including the root).
# We filter to keep only noun-phrase chunks whose label is 'NP'.
np_subtrees = list(tree.subtrees(filter=lambda t: t.label() == 'NP'))
print("\nNumber of NP subtrees:", len(np_subtrees))

print("NLTK Noun Phrases:")
for i, subtree in enumerate(np_subtrees, start=1):
    # What does leaves() return?
    # subtree.leaves() returns a list of terminal nodes (word, POS) tuples for this subtree.
    leaf_tuples = subtree.leaves()
    words_only = [word for word, tag in leaf_tuples]

    print(f"{i}. subtree object: {subtree}")
    print(f"   leaves(): {leaf_tuples}")
    print(f"   noun phrase text: {' '.join(words_only)}")
```

`NLTK`:
- It is the Best for Customization
- It is the most complex but offers the most control.
- We manually define rules, making it perfect for specific linguistic research.
- It is generally the slowest due to its manual, step-by-step nature.

**Weakness**: Very "wordy" code; requires a lot of manual configuration to get accurate results:  explicit, step-by-step, rule-driven architecture. Unlike other libraries that use pre-trained statistical models

- ### Using `spaCy`:
  
`spaCy` uses a pre-trained statistical model. It automatically identifies noun chunks based on the grammatical structure of the sentence.


```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

text = "The quick brown fox jumps over the lazy dog near the old bridge."
doc = nlp(text)

# Extract noun chunks
print("spaCy Noun Phrases:")
for chunk in doc.noun_chunks:
    print(f"- {chunk.text}")
```

`spaCy`: 
- It is the  Best for Performance
- It is the fastest and most accurate option.
- It uses advanced "smart" logic (dependency parsing) to understand grammar.
- Ideal for professional, large-scale applications.

**Weakness**: Requires downloading large model files (~12MB to 500MB).

# Named-Entity-Recogintion (NER):

Named entity recognition (NER) is an information extraction task which identifies mentions of various ***named entities*** in unstructured text and classifies them into ***predetermined categories***, such as *person names*, *organizations*, *locations*, *date/time*, *monetary values*, and so forth. They can, for example, help with the classification of news content, content recommentations and search algorithms.

NLTK uses specific tags to categorize the words it finds. Here are the most common ones you will see:
- **PERSON**: People, including fictional characters.
- **ORGANIZATION**: Companies, agencies, institutions.
- **GPE**: Geo-Political Entity (countries, cities, states).
- **LOCATION**: Non-GPE locations (mountains, bodies of water).
- **FACILITY**: Buildings, airports, highways, bridges.
- **DATE**: Absolute or relative dates or periods.

### Steps to Perform NER with NLTK:
1. Import Necessary Libraries: Begin by downloading and importing the essential libraries and modules for NER.

```python
nltk.download('maxent_ne_chunker', force= True) # The Named Entity Recognition (NER) model for NLTK.
nltk.download('words')  # needed by the chunker
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
#import libraries //Using NLTK
import nltk
from nltk import ne_chunk
from nltk import word_tokenize
from nltk import pos_tag
```

 2. Preprocess the text: Tokenization and part-of-speech (POS) tagging are required before NER.
 3. Apply `NLTK`’s NER tagger: Use `ne_chunk` to identify named entities in the text.

```python
text = "John works at Google in New York."

# Step 1: Tokenize
tokens = word_tokenize(text)

# Step 2: POS Tag (Crucial!)
tags = pos_tag(tokens)

# Step 3: NER Chunking
# binary=True is simpler if you only care about whether something is a named entity, not what kind.
ner_tree = ne_chunk(tags, binary=False) 

print(ner_tree) # Subtrees = named entity chunks (one or more tokens).
```

```python
# In order to make the previous output more readable:
for chunk in ner_tree:
    if hasattr(chunk, 'label'):
        entity_name = ' '.join(c[0] for c in chunk) # c[0] takes the word from each (word, pos) tuple.
        entity_type = chunk.label()
        print(f"Entity: {entity_name} | Type: {entity_type}")
```

```python
# Here's another example

# Sample sentence
sentence = "Apple is looking at buying U.K. startup for $1 billion."

# Tokenize and POS tag
tokens = word_tokenize(sentence) 
pos_tags = pos_tag(tokens)

# Perform NER
named_entities = ne_chunk(pos_tags)
print(named_entities)
```

-------------------------------

We can notice a few "quirks" in how NLTK handles this specific sentences 

- The "Apple" Mistake: NLTK often tags Apple as GPE (Geo-Political Entity/Location) or PERSON because it sees a capitalized noun and guesses based on common names. It rarely correctly identifies it as an ORGANIZATION without more context.
- Money: NLTK’s ne_chunk does not group  "`$ 1 billion`" as a single entity. It treats the symbol and numbers as individual parts of speech ($, CD).
- U.K.: `nltk` had given it a Part-of-Speech (POS) tag which is `NNP` but failed to  give it an entity. This is because `NLTK`'s older models sometimes struggle with periods in abbreviations. It might see `U.K.` as three separate things: `U`, `.`, and `K`.

Limitations of Using `NLTK` for NER:

While `NLTK` is a powerful tool for natural language processing, it has several limitations when it comes to named entity recognition:

- Performance: `NLTK` uses classical machine learning or rule-based methods for NER which may not be as accurate as deep learning methods. This means that the precision and recall rates in entity identification will be lower.
- Scalability: `NLTK` can be less efficient when dealing with big data and high-performance applications, thus it is not ideal for use in production environments that demand real-time processing.
- Flexibility: The library does not provide as many configuration and tuning parameters as more sophisticated libraries such as `SpaCy`; that allow users to define their pipelines and models.

```python
# Importing the library and it's language model
import spacy
nlp = spacy.load("en_core_web_sm")

#Read/create a sentence
doc = nlp("Apple is ready to launch new phone worth $10000 in New york time square")

# NER
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
# Importing the library and it's language model
import spacy
nlp = spacy.load("en_core_web_sm")

#Read/create a sentence
doc = nlp("Apple is looking at buying U.K. startup for $1 billion.")

# NER
for ent in doc.ents:
    print(ent.text, ent.label_)
```

Here we notice that `spacy`  overcame the problemes we faced when we used `nltk` and had correctly gave `Apple`, `U.k.` and `$1 billion` their proper NER Tags!

When to use `spaCy` instead of `NLTK`?:
- Accuracy: `NLTK`'s `ne_chunk` is rule-based and often misses complex names; `spaCy` uses statistical models (deep learning).
- Speed: `spaCy` is optimized for production-level "industrial" NLP.
- Granularity: `spaCy` identifies more types, such as `MONEY`, `QUANTITY`,`TIME`, and `PRODUCT`.
- Ease of Use: You don't have to manage separate downloads for tokenizers, taggers, and chunkers.

# Noun-Phrase-Extraction:

Noun phrase (NP) extraction (also called NP-chunking) is an NLP technique that identifies and extracts nouns and their modifiers (adjectives, determiners) from text to capture key information, often using libraries like `SpaCy`, `NLTK`, or `TextBlob`. It transforms raw text into structured data by locating chunks such as "the big house," often focusing on adjective-noun or noun-noun patterns. It is a fundamental task in Natural Language Processing (NLP) used for summarization, information retrieval, and content understanding. 

Noun phrases are POS patterns that include one or multiple nouns.

Common Patterns of NP: Phrases are identified by specific part-of-speech (POS) sequences, such as:

- Adjective + Noun: "Beautiful text"
- Noun + Noun: "Machine Learning"
- Determiner + Noun: "The cat"
- Determiner + Adjective+  Noun: " The lazy dog".



## Example:

- ### Using `TextBlob`:
`TextBlob` uses a pre-trained "FastNpeDiscovery" extractor by default, which means that we don't have to define any grammar rules or POS tags ourselves.

```python
import nltk
from textblob import TextBlob
#Extract noun
blob = TextBlob("John is learning natural language processing")
for np in blob.noun_phrases:
    print(np)
```

Here's a more complex and lengthy text to compare with different libs:

```python
from textblob import TextBlob

text = "The quick brown fox jumps over the lazy dog near the old bridge."
blob = TextBlob(text)

# Extract noun phrases
print("TextBlob Noun Phrases:")
for np in blob.noun_phrases:
    print(f"- {np}")
```

`TextBlob`: 
- It is the Best for Ease of Use.
- It is the simplest to implement with the least amount of code.
- It offers moderate speed and accuracy, acting as a user-friendly "middle ground."
- Ideal for students, beginners, or quick experimental prototypes.


**Weakness**: "Black box" logic; if it gets a phrase wrong, it is difficult to fix without switching to NLTK.


- ### Using `NLTK`:
NLTK requires a manual approach. You must first tag the words with Part-of-Speech (POS) tags and then define a Regular Expression grammar to "chunk" them together.


```python
import nltk
from nltk import pos_tag, word_tokenize, RegexpParser

text = "The quick brown fox jumps over the lazy dog near the old bridge."

# 1. Tokenize and Tag
tokens = word_tokenize(text)
tags = pos_tag(tokens)

# 2. Define Grammar: (Optional Determiner + any number of Adjectives + Noun)
grammar = "NP: {<DT>?<JJ>*<NN>}"

# 3. Create Parser and Parse
parser = RegexpParser(grammar)
tree = parser.parse(tags)

print("NLTK Noun Phrases:")
for subtree in tree.subtrees(filter=lambda t: t.label() == 'NP'):
    print(f"- {' '.join(word for word, tag in subtree.leaves())}")
```

`NLTK`:
- It is the Best for Customization
- It is the most complex but offers the most control.
- We manually define rules, making it perfect for specific linguistic research.
- It is generally the slowest due to its manual, step-by-step nature.

**Weakness**: Very "wordy" code; requires a lot of manual configuration to get accurate results.

- ### Using `spaCy`:
  
`spaCy` uses a pre-trained statistical model. It automatically identifies noun chunks based on the grammatical structure of the sentence.


```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

text = "The quick brown fox jumps over the lazy dog near the old bridge."
doc = nlp(text)

# Extract noun chunks
print("spaCy Noun Phrases:")
for chunk in doc.noun_chunks:
    print(f"- {chunk.text}")
```

`spaCy`: 
- It is the  Best for Performance
- It is the fastest and most accurate option.
- It uses advanced "smart" logic (dependency parsing) to understand grammar.
- Ideal for professional, large-scale applications.

**Weakness**: Requires downloading large model files (~12MB to 500MB).

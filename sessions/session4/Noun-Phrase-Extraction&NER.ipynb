# Noun-Phrase-Extraction:

Noun phrase (NP) extraction (also called NP-chunking) is an NLP technique that identifies and extracts nouns and their modifiers (adjectives, determiners) from text to capture key information, often using libraries like SpaCy, NLTK, or TextBlob. It transforms raw text into structured data by locating chunks such as "the big house," often focusing on adjective-noun or noun-noun patterns. It is a fundamental task in Natural Language Processing (NLP) used for summarization, information retrieval, and content understanding. 

Noun phrases are POS patterns that include one or multiple nouns.
Common Patterns:
Phrases are identified by specific part-of-speech (POS) sequences:
- Adjective + Noun: "Beautiful text"
- Noun + Noun: "Machine Learnint"
- Determiner + Noun: "The cat."
- Determiner + Adjective+  Noun: " The lazy dog".



TextBlob uses a pre-trained "FastNpeDiscovery" extractor by default, which means you don't have to define any grammar rules or POS tags yourself.

```python
import nltk
from textblob import TextBlob
#Extract noun
blob = TextBlob("John is learning natural language processing")
for np in blob.noun_phrases:
    print(np)
```

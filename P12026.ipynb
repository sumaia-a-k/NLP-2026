# Working with Texts

This notebook introduces one of the most common early tasks in NLP: reading text from files.

## Section 1: Reading PDF Text

In this section, we will use the `pypdf` library to open a PDF file and extract text from it.

`pypdf` is a modern Python library for reading and manipulating PDF files. Other libraries also exist, such as `PyPDF2`, but in this notebook we will use `pypdf`.

## Import the Library

Before reading a PDF, we need to import `PdfReader` from `pypdf`.

`PdfReader` is the class responsible for loading a PDF file and giving us access to its pages.

```python
from pypdf import PdfReader
```

## Open the PDF File

We will work with a file named `pdfExample.pdf`.

The file is opened in binary read mode using `"rb"`. PDF files are binary files, so they should not be opened in normal text mode.

In the next cell, `open()` returns a buffered binary stream object. In Python, this is usually shown as `_io.BufferedReader`.

This buffered reader helps Python read binary file data efficiently. `PdfReader` then uses that stream to parse the PDF structure.

```python
pdf_file = open("./pdfExample.pdf", "rb")
print(type(pdf_file))
```

## Create a PDF Reader Object

Now we pass the opened file into `PdfReader`.

This reader object gives access to the PDF metadata and its pages.

```python
reader = PdfReader(pdf_file)
```

## Get the Number of Pages

A PDF is made of pages. We can count how many pages exist by checking the length of `reader.pages`.

```python
number_of_pages = len(reader.pages)
print("Number of pages:", number_of_pages)
```

## Get a Certain Page by Page Number

Pages are accessed by index.

Important: Python uses zero-based indexing, so:

- page `0` means the first page
- page `1` means the second page
- page `2` means the third page

In the next example, we read the first page. You can change the page number if needed.

```python
page_number = 0
selected_page = reader.pages[page_number]
print("Selected page index:", page_number)
```

## Extract Text from the Selected Page

Once we have a page object, we can call `extract_text()` to read the text content from that page.

Keep in mind that PDF text extraction is not always perfect. The quality depends on how the original PDF was created.

```python
page_text = selected_page.extract_text()
print(page_text)
```

## Close the File

Because we opened the file manually with `open()`, we should close it when we are done.

Closing files is a good habit because it releases system resources.

```python
pdf_file.close()
```

## Recommended Safer Version

A safer and more Pythonic way to work with files is to use a `with` block.

This automatically closes the file after the block finishes, even if an error happens.

```python
with open("./pdfExample.pdf", "rb") as pdf_file:
    reader = PdfReader(pdf_file)
    number_of_pages = len(reader.pages)
    page_number = 0
    selected_page = reader.pages[page_number]
    page_text = selected_page.extract_text()

    print("Buffered reader type:", type(pdf_file))
    print("Number of pages:", number_of_pages)
    print("Text from selected page:")
    print(page_text)
```

## Summary

In this section, you learned how to:

- import `PdfReader` from `pypdf`
- open a PDF file in binary mode
- understand the role of the buffered reader
- count the number of pages
- select a page by page number
- extract text from that page
- close the file correctly

This is a strong first step toward working with text data from documents in NLP tasks.

## Section 2: Reading Word Text

In this section, we will use `python-docx` to read text from a Word file.

`python-docx` is a Python library for working with Microsoft Word `.docx` files. It is a good choice when you want to open a Word document, inspect its paragraphs, and extract text.

For this notebook, we will read the file `./wordExample.docx`.

Unlike PDF handling, we do not need to open the file in binary mode ourselves. The `Document(...)` function handles that for us.

## Import the Library

We will import `Document` from `docx`.

`Document` is the main class used to open and work with a `.docx` file.

```python
from docx import Document
```

## Open the Word File

Now we open the Word document using `Document("./wordExample.docx")`.

This loads the document so we can access its paragraphs and other elements.

```python
doc = Document("./wordExample.docx")
print(type(doc))
```

## Extract All Text

A Word document usually contains its main text as a sequence of paragraphs.

We can loop through `doc.paragraphs`, collect each paragraph's text, and then join everything into one string.

This gives us the full document text in a form that is easy to print, search, or preprocess for NLP tasks.

```python
all_paragraph_text = []

for paragraph in doc.paragraphs:
    all_paragraph_text.append(paragraph.text)

document_text = "\n".join(all_paragraph_text)
print(document_text)
```

## Access a Specific Paragraph

Just like a list in Python, `doc.paragraphs` can be accessed by index.

This is useful when you want to inspect one part of the document at a time.

```python
print(doc.paragraphs[0].text)
```

## Section 3: Fetching and Processing JSON API Data with Python

In this section, we will fetch data from a web API and process the JSON response.

This is useful in NLP because many APIs return text data in JSON format.

This section demonstrates how to:

- make an HTTP request to a REST API
- handle a JSON response
- convert JSON into Python data
- extract and display specific fields

We will use:

- `requests` to send the HTTP request
- `json` to print the response in a readable format

```python
import requests
import json
```

## Fetch the API Response and Parse JSON

In the next cell:

- `requests.get(...)` sends a GET request to the API
- `headers={"Accept": "application/json"}` makes it explicit that we want JSON back
- `response.raise_for_status()` stops execution if the request fails
- `response.json()` converts the JSON response into a Python dictionary
- `json.dumps(..., indent=4)` formats the data to make it easier to read

After that, we extract the `fact` and `length` fields from the returned data.

```python
response = requests.get(
    url="https://catfact.ninja/fact",
    headers={"Accept": "application/json"}
)

response.raise_for_status()

fact_data = response.json()
print(json.dumps(fact_data, indent=4))

cat_fact = fact_data["fact"]
cat_fact_length = fact_data["length"]

print("\nCat fact:", cat_fact)
print("Fact length:", cat_fact_length)
```

## Section 4: Collecting Data from HTML

In this section, we will collect data from an HTML page.

This is a common task in NLP and text mining when useful text is published on websites instead of in local files.

We will use:

- `urllib.request` to fetch the web page
- `BeautifulSoup` to parse and navigate the HTML

In this example, we will read the Wikipedia page for Natural Language Processing.

```python
import urllib.request
from bs4 import BeautifulSoup
```

## Fetch the Web Page and Parse the HTML

In the next cell:

- `urllib.request.Request(...)` creates a request object
- the `User-Agent` header makes the request look like a normal browser request
- `urllib.request.urlopen(...)` opens the web page
- `.read()` returns the page content as bytes
- `BeautifulSoup(..., "html.parser")` parses the HTML structure
- `prettify()` formats the HTML so it is easier to inspect

Some websites block requests that do not include a browser-like `User-Agent`. If you get `HTTP Error 403: Forbidden`, adding that header is often the fix.

This section requires an internet connection when you run it.

```python
url = "https://en.wikipedia.org/wiki/Natural_language_processing"
request = urllib.request.Request(
    url,
    headers={
        "User-Agent": "Mozilla/5.0"
    }
)

response = urllib.request.urlopen(request)
html_doc = response.read()

soup = BeautifulSoup(html_doc, "html.parser")

formatted_html = soup.prettify()
print(formatted_html[:1000])

print("\nTitle Tag:", soup.title)
print("Title Text:", soup.title.string)
print("First <a> tag text:", soup.a.string)
print("First <b> tag text:", soup.b.string)
```

## Section 5: Tokenizing Text into Sentences Using NLTK

In this section, we will use NLTK to split a paragraph into individual sentences.

Sentence tokenization is a common first step in NLP because many later tasks work better when text is already broken into sentence-sized pieces.

We will use:

- `nltk` for natural language processing tools
- `sent_tokenize` for sentence splitting

NLTK uses a pretrained tokenizer called Punkt to detect sentence boundaries.

## Download the Required NLTK Data and Import the Tokenizer

Before sentence tokenization works, NLTK needs the Punkt tokenizer data.

In the next cell:

- we import `nltk`
- we import `sent_tokenize`
- we check whether the tokenizer data already exists
- if it does not exist, we download it

This setup usually only needs to happen once per environment.

## Why Do We Need Punkt?

`Punkt` is the pretrained sentence tokenizer model used by NLTK.

We need it because sentence tokenization is more complex than simply splitting text at every period.

For example, a good tokenizer should avoid mistakes with:

- abbreviations such as `Dr.` or `U.S.A.`
- decimal numbers
- punctuation used inside quotes
- normal sentence endings such as `.`, `!`, and `?`

The Punkt model has learned rules and patterns that help NLTK decide where real sentence boundaries are.

Without this resource, `sent_tokenize(...)` does not have the language data it needs to split text correctly.

```python
import nltk
from nltk.tokenize import sent_tokenize

try:
    nltk.data.find("tokenizers/punkt")
except LookupError:
    print("Downloading NLTK Punkt tokenizer data...")
    nltk.download("punkt")
    nltk.download("punkt_tab")
    print("Download complete!")
```

## Tokenize a Paragraph into Sentences

Now we define a sample paragraph and pass it to `sent_tokenize(...)`.

The tokenizer examines punctuation and capitalization patterns to decide where one sentence ends and the next begins.

The result is a Python list where each item is one sentence.

```python
paragraph_text = "Hello World. It's good to see you. Thanks for buying this book."

sentences = sent_tokenize(paragraph_text)

print("Tokenized sentences:")
for index, sentence in enumerate(sentences, 1):
    print(f"{index}. {sentence}")
```

## What Happened Here?

In this example:

- the input was one paragraph string
- `sent_tokenize(...)` split it into separate sentences
- the result was stored in `sentences`
- we printed each sentence with its position number

This kind of sentence splitting is often used before summarization, classification, information extraction, and other NLP tasks.

## Multilingual Sentence Tokenization (Spanish, Russian, Japanese, Arabic)

You can continue from the previous tokenizer section and handle more languages.

- Spanish and Russian usually work with NLTK's Punkt tokenizer.
- Japanese and Arabic may need a simpler fallback if language-specific models are not available.

The next cell uses:

- `sent_tokenize` for languages supported by Punkt
- a punctuation-based fallback for Japanese and Arabic

```python
import re

# One-time setup for sentence tokenization models
nltk.download("punkt")
nltk.download("punkt_tab")

def simple_sentence_split(text):
    # Fallback splitter for languages where Punkt is not reliable.
    parts = re.split(r"(?<=[.!?؟。！？])\s*", text.strip())
    return [part for part in parts if part]

def tokenize_multilingual(text, language):
    punkt_languages = {
        "english", "spanish", "russian", "french", "german", "italian",
        "portuguese", "dutch", "danish", "swedish", "norwegian", "finnish",
        "turkish", "greek", "czech", "polish", "slovene", "estonian"
    }

    if language in punkt_languages:
        return sent_tokenize(text, language=language)

    if language in {"japanese", "arabic"}:
        return simple_sentence_split(text)

    raise ValueError(f"Unsupported language: {language}")

print("Spanish:", tokenize_multilingual("Hola mundo. ¿Como estas? Todo bien.", "spanish"))
print("Russian:", tokenize_multilingual("Привет мир. Как дела? Все хорошо.", "russian"))
print("Japanese:", tokenize_multilingual("こんにちは世界。元気ですか？私は元気です。", "japanese"))
print("Arabic:", tokenize_multilingual("مرحبا بالعالم. كيف حالك؟ انا بخير.", "arabic"))
```

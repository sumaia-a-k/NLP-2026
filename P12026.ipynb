# Working with Texts

This notebook introduces one of the most common early tasks in NLP: reading text from files.

## Section 1: Reading PDF Text

In this section, we will use the `pypdf` library to open a PDF file and extract text from it.

`pypdf` is a modern Python library for reading and manipulating PDF files. Other libraries also exist, such as `PyPDF2`, but in this notebook we will use `pypdf`.

## Import the Library

Before reading a PDF, we need to import `PdfReader` from `pypdf`.

`PdfReader` is the class responsible for loading a PDF file and giving us access to its pages.

```python
from pypdf import PdfReader
```

## Open the PDF File

We will work with a file named `pdfExample.pdf`.

The file is opened in binary read mode using `"rb"`. PDF files are binary files, so they should not be opened in normal text mode.

In the next cell, `open()` returns a buffered binary stream object. In Python, this is usually shown as `_io.BufferedReader`.

This buffered reader helps Python read binary file data efficiently. `PdfReader` then uses that stream to parse the PDF structure.

```python
pdf_file = open("./pdfExample.pdf", "rb")
print(type(pdf_file))
```

## Create a PDF Reader Object

Now we pass the opened file into `PdfReader`.

This reader object gives access to the PDF metadata and its pages.

```python
reader = PdfReader(pdf_file)
```

## Get the Number of Pages

A PDF is made of pages. We can count how many pages exist by checking the length of `reader.pages`.

```python
number_of_pages = len(reader.pages)
print("Number of pages:", number_of_pages)
```

## Get a Certain Page by Page Number

Pages are accessed by index.

Important: Python uses zero-based indexing, so:

- page `0` means the first page
- page `1` means the second page
- page `2` means the third page

In the next example, we read the first page. You can change the page number if needed.

```python
page_number = 0
selected_page = reader.pages[page_number]
print("Selected page index:", page_number)
```

## Extract Text from the Selected Page

Once we have a page object, we can call `extract_text()` to read the text content from that page.

Keep in mind that PDF text extraction is not always perfect. The quality depends on how the original PDF was created.

```python
page_text = selected_page.extract_text()
print(page_text)
```

## Close the File

Because we opened the file manually with `open()`, we should close it when we are done.

Closing files is a good habit because it releases system resources.

```python
pdf_file.close()
```

## Recommended Safer Version

A safer and more Pythonic way to work with files is to use a `with` block.

This automatically closes the file after the block finishes, even if an error happens.

```python
with open("./pdfExample.pdf", "rb") as pdf_file:
    reader = PdfReader(pdf_file)
    number_of_pages = len(reader.pages)
    page_number = 0
    selected_page = reader.pages[page_number]
    page_text = selected_page.extract_text()

    print("Buffered reader type:", type(pdf_file))
    print("Number of pages:", number_of_pages)
    print("Text from selected page:")
    print(page_text)
```

## Summary

In this section, you learned how to:

- open a PDF file in binary mode
- understand the role of the buffered reader
- count the number of pages
- select a page by page number
